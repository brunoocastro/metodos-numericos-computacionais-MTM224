```python
# Determine uma raiz não-nula com epsilon <= 0.01
import numpy as np
from lib.Plotter import Plotter


epsilon = 0.001

q_aula_10_04 = lambda x: x + 1 - (2 * np.sin(np.pi * x))

q_aula_plot = Plotter(q_aula_10_04)

print("Função a ser avaliada. Intervalo [-1,1]")
q_aula_plot.plot_functions(-1, 1)
```

```python
f = q_aula_10_04 # Pra facilitar

def validateInterval(a,b):
  print(f"f(a) = {f(a)} e f(b) = {f(b)}")
  produto = f(a) * f(b)
  print(f"{produto} < 0 ? {'Sim, existe pelo menos uma raiz no intervalo.' if produto < 0 else 'Não, intervalo inválido'}")
```

```python
# Teste de intervalos 1:
a = -2
b = 2

validateInterval(a,b)
```

```python
# Teste de intervalos 2:
a = 0
b = 2

validateInterval(a,b)
```
