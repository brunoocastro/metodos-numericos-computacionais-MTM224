```python
from lib.zeros_funcoes.newton_raphson import NewtonRaphson
import numpy as np
```

```python
f = lambda x: 2*x * np.cos(2*x) - (x+1)**2
fprime = lambda x: 2*np.cos(2*x) - 4*x * np.sin(2*x) - 2*(x+1)
max_iterations = 1000
epsilon = 1e-4
x0 = 0.30
```

```python
alg = NewtonRaphson(f, fprime, x0, epsilon, max_iterations)
```

```python
alg.calculate_root()
```
