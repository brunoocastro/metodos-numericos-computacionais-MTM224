```python
import numpy as np
from lib.zeros_funcoes.falsa_posicao import FalsaPosicao
```

```python
f_bernardo = lambda x: x * np.cos(x) - 2 * x**2 + 3*x - 1

a = 0
b = np.pi / 4
epsilon = 0.001
res = FalsaPosicao(f_bernardo, a, b, epsilon)

res.calculate_root()
res.show_final_table()
```

```python
faula = lambda x: 4*np.cos(x) - np.exp(2*x)
epsilon = 0.001
a = 0.5
b = 0.6

falsapos = FalsaPosicao(faula, a, b, epsilon)
falsapos.calculate_root()
falsapos.show_final_table()
```
